This notebook runs Mamba on different datasets, using the official implementation from:
https://github.com/state-spaces/mamba

```python
"""
How to set up the Mamba environment for running experiments
conda env create -f mamba_environment.yml
conda activate mamba_env
"""
from Mamba_utils import run_experiment
from config import ALL_CONFIGS

# This dictionary will store the detailed results from all experiments
experiment_results = {}
```

```python
"""
Main script to run Mamba experiments with different configurations.
This script iterates over all configurations defined in the config module.
So for each configuration initializes the dataset, the model, the optimizer, and the loss function, 
and then runs the training and evaluation processes.
This step needs approximately 10 minutes to complete.
"""
for config in ALL_CONFIGS:
    run_experiment(config, experiment_results)
```

```python
""" Print a summary of all experiments run"""

print(f"\n{'='*85}\n Final Summary of All Experiments \n{'='*85}")

if not experiment_results:
    print("No experiment results were found. Please run the experiment cell above.")
else:
    print(f"{'Experiment Name':<30} | {'d_model':<10} | {'Final Val Accuracy':<20} | {'Final Test Accuracy':<20}")
    print("-" * 85)
    
    for name, results in experiment_results.items():
        val_acc = results.get('history', {}).get('val_accuracy', [None])[-1]
        test_acc = results.get('test_accuracy', None)
        d_model = results.get('config', {}).get('d_model', 'N/A')

        val_acc_str = f"{val_acc:.2f}%" if val_acc is not None else "N/A"
        test_acc_str = f"{test_acc:.2f}%" if test_acc is not None else "N/A"
        
        print(f"{name:<30} | {str(d_model):<10} | {val_acc_str:<20} | {test_acc_str:<20}")
```

## Pretrained Mamba experiments
Run pretrained Mamba (e.g., `state-spaces/mamba-2.8b`) side-by-side.


```python
from Mamba_utils import run_experiment_pretrained
```

```python
""" Run pretrained Mamba for a side-by-side comparison """
from copy import deepcopy
from config import ALL_CONFIGS

PRETRAINED_MODEL = "state-spaces/mamba-2.8b"  # e.g., use "state-spaces/mamba-130m" for lighter runs
SELECTED_CONFIGS = [ALL_CONFIGS]

PRETRAINED_CONFIGS = []
for cfg in SELECTED_CONFIGS:
    pre_cfg = deepcopy(cfg)
    pre_cfg.update({
        "experiment_name": f"{cfg['experiment_name']}_pretrained",
        "pretrained_model_name": PRETRAINED_MODEL,
        "freeze_backbone": True,
        "epochs": min(5, pre_cfg.get("epochs", 5)),
        "batch_size": min(8, pre_cfg.get("batch_size", 8)),
        "use_amp": True,
        # "torch_dtype": "bfloat16",  # uncomment if supported
        # "device_map": "auto",        # requires accelerate, sharded loading
    })
    PRETRAINED_CONFIGS.append(pre_cfg)

for config in PRETRAINED_CONFIGS:
    run_experiment_pretrained(config, experiment_results)
```

```python
""" Re-run summary to include pretrained results """

print(f"\n{'='*85}\n Final Summary (Including Pretrained) \n{'='*85}")

if not experiment_results:
    print("No experiment results were found. Please run the experiment cells above.")
else:
    print(f"{'Experiment Name':<30} | {'d_model':<10} | {'Final Val Accuracy':<20} | {'Final Test Accuracy':<20}")
    print("-" * 85)
    for name, results in experiment_results.items():
        val_acc = results.get('history', {}).get('val_accuracy', [None])[-1]
        test_acc = results.get('test_accuracy', None)
        d_model = results.get('config', {}).get('d_model', 'N/A')
        val_acc_str = f"{val_acc:.2f}%" if val_acc is not None else "N/A"
        test_acc_str = f"{test_acc:.2f}%" if test_acc is not None else "N/A"
        print(f"{name:<30} | {str(d_model):<10} | {val_acc_str:<20} | {test_acc_str:<20}")
```
